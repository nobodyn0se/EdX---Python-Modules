```python
def char_art(steps):
    for row in range(steps + 1):
        for col in range(steps + 1):
            if(row == col):
                print(col * "[]")
            
char_art(63)
```

```python
table = [[5, 2, 6], [4, 6, 0], [9, 1, 8], [7, 3, 8]]

for row in table:
    for col in row:
        # print the value col followed by a tab
        print(col, end = "\t")

    # Print a new line
    print()
```
